```python
# ==============================================
# ACTIVIDAD: ANÁLISIS ORGANIZACIONAL CON PYTHON
# ==============================================

# Instalar librerías necesarias
!pip install faker vaderSentiment

# Importar librerías
import random
import numpy as np
import pandas as pd
from faker import Faker
from datetime import datetime
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from google.colab import files

# Configuración visual
sns.set(style="whitegrid")

# ==========================================
# 1. GENERACIÓN DE BASE DE DATOS SINTÉTICA
# ==========================================

fake = Faker()
Faker.seed(123)
np.random.seed(123)
random.seed(123)

n = 1200

departments = ['Ventas', 'Soporte', 'Ingeniería', 'Marketing', 'RRHH', 'Operaciones', 'Finanzas']

roles = {
    'Ventas': ['Ejecutivo','Gerente','Asistente'],
    'Soporte': ['Analista','Ingeniero','Coordinador'],
    'Ingeniería': ['Desarrollador','QA','Arquitecto'],
    'Marketing': ['Especialista','Content','Manager'],
    'RRHH': ['Generalista','Recruiter','Manager'],
    'Operaciones': ['Supervisor','Coordinador','Analista'],
    'Finanzas': ['Contador','Analista','Controller']
}

cities = [
    ('Ciudad de México', 19.4326, -99.1332),
    ('Monterrey', 25.6866, -100.3161),
    ('Guadalajara', 20.6597, -103.3496),
    ('Querétaro', 20.5888, -100.3899),
    ('Puebla', 19.0413, -98.2062),
    ('Tijuana', 32.5149, -117.0382),
    ('Mérida', 20.9674, -89.5926)
]

positive_phrases = [
    "Me siento apoyado por mi jefe",
    "Buen balance vida-trabajo",
    "Reconocen mi esfuerzo",
    "Capacitación útil",
    "Equipo colaborativo"
]

negative_phrases = [
    "Sobrecarga de trabajo",
    "Pago insuficiente",
    "Falta de liderazgo",
    "Procesos confusos",
    "Pocas oportunidades de crecimiento"
]

neutral_phrases = [
    "Tareas rutinarias",
    "Horario estable",
    "Herramientas adecuadas"
]

rows = []

for i in range(1, n + 1):
    dept = random.choice(departments)
    role = random.choice(roles[dept])
    city, lat, lon = random.choice(cities)

    hire_date = fake.date_between(start_date='-8y', end_date='today')
    tenure_months = (datetime.today().year - hire_date.year) * 12 + (datetime.today().month - hire_date.month)

    dept_base = {
        'Ventas': 6.0,
        'Soporte': 5.5,
        'Ingeniería': 7.0,
        'Marketing': 6.5,
        'RRHH': 7.0,
        'Operaciones': 5.8,
        'Finanzas': 6.2
    }[dept]

    satisfaction = max(1, min(10, np.round(np.random.normal(dept_base, 1.6), 1)))

    workload = max(
        20,
        int(np.random.normal(40 if dept in ['Ventas', 'Operaciones'] else 38, 8))
    )

    performance = int(np.clip(np.round(np.random.normal(3.5 + (satisfaction - 6) / 3, 0.9)), 1, 5))

    remote = 1 if dept in ['Ingeniería', 'Marketing', 'RRHH'] and random.random() < 0.6 else \
             (1 if random.random() < 0.12 else 0)

    incidents = int(np.random.poisson(0.12 * workload / 10))

    training_hours = int(max(0, np.random.normal(12 if dept in ['Ingeniería', 'RRHH', 'Marketing'] else 6, 5)))

    prob_attr = 0.02 + (6 - satisfaction) * 0.05 + max(0, (workload - 45)) * 0.002 - training_hours * 0.004
    prob_attr = np.clip(prob_attr, 0.01, 0.6)

    attrition = 1 if random.random() < prob_attr else 0

    p = random.random()

    if p < 0.45:
        feedback = random.choice(positive_phrases)
    elif p < 0.8:
        feedback = random.choice(neutral_phrases)
    else:
        feedback = random.choice(negative_phrases)

    if random.random() < 0.15:
        feedback += ". " + random.choice(positive_phrases + neutral_phrases + negative_phrases)

    perf_note = random.choice(["Cumple", "Excede", "Necesita mejorar", "Sobresaliente"])

    row = {
        'employee_id': f"E{10000+i}",
        'department': dept,
        'role': role,
        'city': city,
        'lat': lat + np.random.normal(0, 0.02),
        'lon': lon + np.random.normal(0, 0.02),
        'hire_date': hire_date,
        'tenure_months': tenure_months,
        'satisfaction': satisfaction,
        'workload_hours_per_week': workload,
        'performance_score': performance,
        'remote': remote,
        'incidents_last_year': incidents,
        'training_hours': training_hours,
        'attrition': attrition,
        'feedback_text': feedback,
        'performance_note': perf_note
    }

    rows.append(row)

df = pd.DataFrame(rows)

print("Base de datos generada:")
display(df.head())

# =======================
# 2. VARIABLES DERIVADAS
# =======================

df['tenure_years'] = (df['tenure_months'] / 12).round(1)

df['overloaded'] = df['workload_hours_per_week'].apply(
    lambda x: 1 if x > 50 else 0
)

df['remote_label'] = df['remote'].apply(
    lambda x: 'Remoto' if x == 1 else 'Presencial'
)

df['attrition_label'] = df['attrition'].apply(
    lambda x: 'Salió' if x == 1 else 'Permanece'
)

# ============================
# 3. ANÁLISIS DE SENTIMIENTOS
# ============================

analyzer = SentimentIntensityAnalyzer()

df['sentiment_score'] = df['feedback_text'].apply(
    lambda texto: analyzer.polarity_scores(str(texto))['compound']
)

def clasificar_sentimiento(score):
    if score >= 0.05:
        return 'Positivo'
    elif score <= -0.05:
        return 'Negativo'
    else:
        return 'Neutral'

df['sentiment_label'] = df['sentiment_score'].apply(clasificar_sentimiento)

display(df[['employee_id', 'feedback_text', 'sentiment_score', 'sentiment_label']].head(10))

# ===================
# 4. RESUMEN GENERAL
# ===================

print("Tamaño de la base de datos:", df.shape)
print("\nValores faltantes:")
print(df.isnull().sum())

print("\nEstadísticas generales:")
display(df.describe())

# ==========================================
# 5. RESUMEN POR DEPARTAMENTO CON CICLO FOR
# ==========================================

dept_summary = []

for dept in df['department'].unique():
    subset = df[df['department'] == dept]

    avg_sat = subset['satisfaction'].mean()
    avg_workload = subset['workload_hours_per_week'].mean()
    avg_attrition = subset['attrition'].mean()
    avg_training = subset['training_hours'].mean()
    avg_sentiment = subset['sentiment_score'].mean()

    dept_summary.append([
        dept,
        round(avg_sat, 2),
        round(avg_workload, 2),
        round(avg_attrition, 3),
        round(avg_training, 2),
        round(avg_sentiment, 3)
    ])

dept_df = pd.DataFrame(
    dept_summary,
    columns=[
        'Departamento',
        'Satisfacción promedio',
        'Carga promedio',
        'Rotación promedio',
        'Capacitación promedio',
        'Sentimiento promedio'
    ]
)

print("Resumen por departamento:")
display(dept_df)

# =============================
# 6. COMPROBACIÓN DE HIPÓTESIS
# =============================

print("HIPÓTESIS 1: Mayor carga de trabajo = menor satisfacción")

over = df[df['overloaded'] == 1]
not_over = df[df['overloaded'] == 0]

print("Satisfacción promedio de sobrecargados:", round(over['satisfaction'].mean(), 2))
print("Satisfacción promedio de no sobrecargados:", round(not_over['satisfaction'].mean(), 2))

print("\nHIPÓTESIS 2: Más capacitación = menor rotación")

low_training = df[df['training_hours'] <= 5]
high_training = df[df['training_hours'] > 10]

print("Rotación con baja capacitación:", round(low_training['attrition'].mean(), 3))
print("Rotación con alta capacitación:", round(high_training['attrition'].mean(), 3))

print("\nHIPÓTESIS 3: Comentarios negativos = menor satisfacción")

sentiment_group = df.groupby('sentiment_label')['satisfaction'].mean().reset_index()
display(sentiment_group)

print("\nHIPÓTESIS 4: Empleados remotos = mayor satisfacción")

remote_group = df.groupby('remote_label')['satisfaction'].mean().reset_index()
display(remote_group)

# ============
# 7. GRÁFICAS
# ============

# GRÁFICA 1: Distribución de satisfacción
plt.figure(figsize=(7, 4))
sns.histplot(df['satisfaction'], bins=20, kde=True)
plt.title('Gráfica 1. Distribución de la satisfacción laboral')
plt.xlabel('Satisfacción')
plt.ylabel('Frecuencia')
plt.show()

# GRÁFICA 2: Satisfacción vs sobrecarga
plt.figure(figsize=(7, 4))
sns.boxplot(x='overloaded', y='satisfaction', data=df)
plt.xticks([0, 1], ['No sobrecargado', 'Sobrecargado'])
plt.title('Gráfica 2. Satisfacción según sobrecarga laboral')
plt.xlabel('Condición de carga')
plt.ylabel('Satisfacción')
plt.show()

# GRÁFICA 3: Rotación por departamento
plt.figure(figsize=(9, 4))
rot_dept = df.groupby('department')['attrition'].mean().sort_values(ascending=False).reset_index()
sns.barplot(data=rot_dept, x='department', y='attrition')
plt.title('Gráfica 3. Tasa de rotación por departamento')
plt.xlabel('Departamento')
plt.ylabel('Tasa de rotación')
plt.xticks(rotation=45)
plt.show()

# GRÁFICA 4: Capacitación vs rotación
train_bins = pd.cut(df['training_hours'], bins=[-1, 5, 10, 20, 50, 100])
agg_training = df.groupby(train_bins)['attrition'].mean().reset_index()
agg_training['training_hours'] = agg_training['training_hours'].astype(str)

plt.figure(figsize=(8, 4))
sns.barplot(x='training_hours', y='attrition', data=agg_training)
plt.title('Gráfica 4. Rotación según horas de capacitación')
plt.xlabel('Rango de horas de capacitación')
plt.ylabel('Tasa de rotación')
plt.xticks(rotation=45)
plt.show()

# GRÁFICA 5: Satisfacción vs desempeño
plt.figure(figsize=(7, 4))
sns.scatterplot(
    data=df,
    x='satisfaction',
    y='performance_score',
    hue='remote_label',
    alpha=0.7
)
plt.title('Gráfica 5. Satisfacción vs desempeño')
plt.xlabel('Satisfacción')
plt.ylabel('Desempeño')
plt.show()

# GRÁFICA 6: Sentimiento del feedback
plt.figure(figsize=(7, 4))
sns.countplot(data=df, x='sentiment_label', order=['Positivo', 'Neutral', 'Negativo'])
plt.title('Gráfica 6. Clasificación de sentimientos en comentarios')
plt.xlabel('Sentimiento')
plt.ylabel('Cantidad de comentarios')
plt.show()

# GRÁFICA 7: Heatmap de correlación
corr = df[
    [
        'satisfaction',
        'workload_hours_per_week',
        'performance_score',
        'training_hours',
        'attrition',
        'sentiment_score',
        'incidents_last_year'
    ]
].corr()

plt.figure(figsize=(9, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Gráfica 7. Matriz de correlación entre variables')
plt.show()

# GRÁFICA 8: Mapa de colaboradores por ciudad
fig = px.scatter_mapbox(
    df,
    lat='lat',
    lon='lon',
    color='department',
    hover_name='employee_id',
    hover_data=['city', 'satisfaction', 'workload_hours_per_week', 'sentiment_label'],
    zoom=4,
    height=500,
    title='Gráfica 8. Distribución geográfica de colaboradores'
)

fig.update_layout(mapbox_style='open-street-map')
fig.show()

# ===============================================================
# 8. IDENTIFICACIÓN DE EMPLEADOS CRÍTICOS CON CONDICIONES Y CICLO
# ===============================================================

criticos = []

#for i, row in df.iterrows():
#    if row['satisfaction'] <= 4 and row['workload_hours_per_week'] > 45 and row['attrition'] == 0:
#        criticos.append(row['employee_id'])

for i, row in df.iterrows():
    if row['satisfaction'] <= 4 and row['overloaded'] == 1 and row['attrition'] == 0:
        criticos.append(row['employee_id'])


print("Empleados críticos identificados:", len(criticos))
print(criticos[:20])

df_criticos = df[df['employee_id'].isin(criticos)]

display(
    df_criticos[
        [
            'employee_id',
            'department',
            'city',
            'satisfaction',
            'workload_hours_per_week',
            'training_hours',
            'feedback_text',
            'sentiment_label'
        ]
    ].head(20)
)

# ======================================
# 9. DATO INTELIGENTE: ÍNDICE DE RIESGO
# ======================================

df['risk_index'] = (
    (10 - df['satisfaction']) * 0.35 +
    df['overloaded'] * 2 +
    df['attrition'] * 3 +
    (df['sentiment_score'] < 0).astype(int) * 1.5 +
    (df['training_hours'] < 5).astype(int) * 1
)

def clasificar_riesgo(x):
    if x >= 6:
        return 'Alto'
    elif x >= 3:
        return 'Medio'
    else:
        return 'Bajo'

df['risk_level'] = df['risk_index'].apply(clasificar_riesgo)

print("Distribución del nivel de riesgo:")
display(df['risk_level'].value_counts())

plt.figure(figsize=(7, 4))
sns.countplot(data=df, x='risk_level', order=['Bajo', 'Medio', 'Alto'])
plt.title('Gráfica 9. Nivel de riesgo de colaboradores')
plt.xlabel('Nivel de riesgo')
plt.ylabel('Cantidad')
plt.show()

# =======================
# 10. EXPORTAR ARCHIVOS
# =======================

df.to_csv('org_data_synthetic_analizado.csv', index=False, encoding='utf-8')
dept_df.to_csv('resumen_departamentos.csv', index=False, encoding='utf-8')
df_criticos.to_csv('empleados_criticos.csv', index=False, encoding='utf-8')

files.download('org_data_synthetic_analizado.csv')
files.download('resumen_departamentos.csv')
files.download('empleados_criticos.csv')

# ==============================
# 11. CONCLUSIONES AUTOMÁTICAS
# ==============================

print("\nCONCLUSIONES DEL ANÁLISIS")
print("1. Se generó una base de datos sintética con información organizacional de 1,200 colaboradores.")
print("2. Se analizaron variables como satisfacción, carga laboral, desempeño, capacitación, rotación y sentimiento del feedback.")
print("3. La sobrecarga laboral se evaluó como posible factor asociado a menor satisfacción.")
print("4. Las horas de capacitación se analizaron como posible variable relacionada con menor rotación.")
print("5. El análisis de sentimientos permitió convertir comentarios cualitativos en datos cuantitativos.")
print("6. Se identificaron colaboradores críticos usando condiciones y ciclos.")
print("7. El índice de riesgo integra varias variables y funciona como dato inteligente para apoyar la toma de decisiones.")
```
